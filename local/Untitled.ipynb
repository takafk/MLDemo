```python
from datetime import datetime as dt

def to_datetime(df):
    
    tstr = str(df['<DTYYYYMMDD>']) + str(df['<TIME>'])
    
    return dt.strptime(tstr, '%Y%m%d%H%M%S')
```

```python
df['timestamp'] = df.apply(to_datetime, axis=1)
```
